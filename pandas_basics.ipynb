## Cleaning Data

```python
# No matter what data set you are going to work with, you will have some missing data.. 
## So you would need to work on fixing it
import numpy as np

people = {
    'first': ['John', 'Bill', 'Jane', 'Amy', np.nan, None, 'NA'],
    'last': ['Doe', 'Gates', 'Austen', 'Frank', np.nan, np.nan, 'Missing'],
    'email': ['johndoe@gmail.com', 'billgates@gmail', 'janeausten@gmail.com', None, np.nan, 'anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df_cleaning = pd.DataFrame(people)
```

```python
df_cleaning
```

```python
## drop NA
df_cleaning.dropna()
```

```python
## dropna( ) by default is dropna(axis='index', how='any')
df_cleaning.dropna(axis='index', how='any')
# same result as above
```

```python
## axis ='index' --> drops rows if they have missing values
## if 'columns', then will drop columns if they have missing values
## how --> criteria on how to drop --> with 'any', drops the rows with any missing values
## may be not what we want always. MAy be can live with missing emails but may be only want to drop if all 
## the values are null
df_cleaning.dropna(axis='index', how='all')
```

```python
## Get more rows than before
```

```python
## If we change to columns wont delete coz none of the columns have all missing values
df_cleaning.dropna(axis='columns', how='all')
```

```python
## but if we change to any --> any single value is none, then the column dropped
df_cleaning.dropna(axis='columns', how='any')
# empty df -- coz every col has atleast one missing
```

```python
# rebuild the df again for the below
df_cleaning.columns
```

```python
## may be we only want to drop rows where only certain columns have missing values. 
## Can pass a subset of column names

df_cleaning.dropna(axis='index', how='any', subset=['email'])
```

```python
## So results returned when atleast the email value is filled
# this NA is a string - custom missing value

# how really doesnt matter here
```
