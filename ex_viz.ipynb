```python
import matplotlib.pyplot as plt
```

```python
frequency1 = [5,20,50,100,200]
PSD1 = [0.09, 0.09, 0.0007, 0.0007, 0.0007]
label1 = "vibe Event Z"


frequency2 = [5,20,50,100,200]
PSD2 = [0.1, 0.1, 0.0015, 0.0006, 0.0006]
label2 = "vibe Event Y"


frequency3 = [5,20,50,100,200]
PSD3 = [0.03, 0.03, 0.0004, 0.0002, 0.0002]
label3 = "vibe Event X"
```

```python
plt_list = []
plt_list.append([frequency1, PSD1, label1])
plt_list.append([frequency2, PSD2, label2])
plt_list.append([frequency3, PSD3, label3])
```

```python
def grms_plot(plots_list):
    for pl in plots_list:
        freq = pl[0]
        psd = pl[1]
        plt_label = pl[2]
        plt.plot(freq, psd, label=plt_label)
        plt.legend()
    plt.show()
```

```python
# print(plt_list)

grms_plot(plt_list)
```
